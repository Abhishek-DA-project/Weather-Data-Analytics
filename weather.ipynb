```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Dataset**

```python
df = pd.read_csv('weather.csv')

print(df.head())
print(df.info())
```

**CLEAN THE DATA**

```python
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

# **Statistical Analysis**

```python
print(df.describe())
```

```python
print(np.mean(df['MaxTemp']))
print(np.median(df['Rainfall']))
print(np.std(df['Humidity3pm']))
```

***Temperature Distribution***

```python
plt.figure(figsize=(8,5))
plt.hist(df['MaxTemp'], bins=20 ,color = "purple" ,edgecolor = "black")
plt.title('Maximum Temperature Distribution' , color = 'red')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

***Rainfall Analysis***

```python
plt.figure(figsize=(10,5))
plt.plot(df['Rainfall'])
plt.title('Rainfall Trend' , color = "purple" ,size = 20 )
plt.xlabel('Days')
plt.ylabel('Rainfall')
plt.show()
```

***Humidity vs Temperature***

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Humidity3pm', y='Temp3pm', data=df)
plt.title('Humidity vs Temperature')
plt.show()
```

***Heatmap***

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

***Rain Prediction***

```python
sns.countplot(x='RainTomorrow', data=df)
plt.title('Rain Tomorrow Prediction Count')
plt.show()
```
